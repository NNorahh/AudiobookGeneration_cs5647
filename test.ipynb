```python
def word_to_char_offsets(context):
    """
    将 context 中每个单词的字符位置计算出来。
    返回：[(start, end), ...] 表示每个单词的字符范围。
    """
    words = context.split()  # 按空格分词
    offsets = []
    current_pos = 0

    for word in words:
        # 查找当前单词的起始位置，并更新结束位置
        start = context.find(word, current_pos)
        end = start + len(word)
        offsets.append((start, end))
        current_pos = end  # 移动指针，避免重复匹配

    return offsets

def convert_word_to_char_indices(context, word_positions):
    """
    将数据集中的单词级别索引转换为字符级别索引。
    """
    char_offsets = word_to_char_offsets(context)  # 获取每个单词的字符范围

    # 遍历所有单词级别位置，转换为字符级别位置
    char_indices = [
        (char_offsets[start][0], char_offsets[end][1])
        for (start, end) in word_positions
    ]

    return char_indices

# 示例数据
context = "Henry Tilney is a charming character in the story."
target_dialogue = [(0, 1)]  # 单词级别： 'Henry Tilney'
speaker_positions = [(0, 0), (5, 5)]  # 单词级别：'Henry' 和 'character'

# 转换为字符级别
target_dialogue_char = convert_word_to_char_indices(context, target_dialogue)
speaker_positions_char = convert_word_to_char_indices(context, speaker_positions)

# 打印结果
print(f"Target Dialogue (Char Level): {target_dialogue_char}")
print(f"Speaker Positions (Char Level): {speaker_positions_char}")
```

```python
import json

def center_context_by_target(context, target_dialogue, speaker_positions, max_length=512):
    # 将 context 拆分为单词列表
    words = context.split()
    context_length = len(words)

    # 提取 target_dialogue 的单词索引范围
    target_start, target_end = target_dialogue
    center = (target_start + target_end) // 2

    # 计算裁剪的起点和终点，以 target_dialogue 为中心
    half_window = max_length // 2
    start = max(0, center - half_window)
    end = min(context_length, center + half_window)

    # 如果裁剪后的长度不足 max_length，尝试向左右填充
    if end - start < max_length:
        if start == 0:
            end = min(context_length, start + max_length)
        elif end == context_length:
            start = max(0, end - max_length)

    # 裁剪后的上下文
    centered_context = " ".join(words[start:end])

    # 更新 target_dialogue 和 speaker_positions 的索引范围
    new_target_start = target_start - start
    new_target_end = target_end - start

    # 过滤并更新超出裁剪范围的 speaker_positions
    new_speaker_positions = [
        [sp_start - start, sp_end - start]
        for sp_start, sp_end in speaker_positions
        if start <= sp_start < end and start <= sp_end < end
    ]

    return centered_context, [new_target_start, new_target_end], new_speaker_positions


def process_dataset(input_path, output_path, max_length=512):
    """
    处理数据集：裁剪上下文、更新 target_dialogue 和 speaker_positions，并删除无效项。
    """
    with open(input_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    processed_data = []

    for item in data:
        context = item["context"]
        target_dialogue = item["target_dialogue"]
        speaker_positions = item["speaker_position"]

        # 裁剪上下文，并更新 target_dialogue 和 speaker_positions 的范围
        new_context, new_target_dialogue, new_speaker_positions = center_context_by_target(
            context, target_dialogue, speaker_positions, max_length
        )
        
        if new_target_dialogue[0] < 0:
            continue

        # 跳过没有有效 speaker_position 的条目
        if not new_speaker_positions:
            continue

        # 更新处理后的数据
        processed_item = {
            "context": new_context,
            "target_dialogue": new_target_dialogue,
            "speaker_position": new_speaker_positions
        }
        processed_data.append(processed_item)

    # 保存处理后的数据集
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(processed_data, f, ensure_ascii=False, indent=4)

# 输入输出文件路径
input_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\val_raw.json"
output_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\processed_val.json"

# 运行数据处理
process_dataset(input_path, output_path)
```

```python
import json
import os

def word_to_char_offsets(context):
    """
    将 context 中每个单词的字符位置计算出来。
    返回：[(start, end), ...] 表示每个单词的字符范围。
    """
    words = context.split()  # 按空格分词
    offsets = []
    current_pos = 0

    for word in words:
        start = context.find(word, current_pos)
        end = start + len(word)
        offsets.append((start, end))
        current_pos = end

    return offsets

def convert_word_to_char_indices(context, word_positions):
    """
    将单词级别索引转换为字符级别索引。
    """
    char_offsets = word_to_char_offsets(context)
    # print(char_offsets)
    # print(word_positions)
    char_indices = [
        (char_offsets[start][0], char_offsets[end][1])
        for (start, end) in word_positions
    ]
    return char_indices

def process_dataset(file_path):
    """
    批量处理数据集，将 target_dialogue 和 speaker_position 转换为字符级别。
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for entry in data:
        context = entry['context']
        
        # 转换 target_dialogue 和 speaker_position
        entry['target_dialogue'] = convert_word_to_char_indices(
            context, [entry['target_dialogue']]
        )[0]  # 展平单个对话位置

        entry['speaker_position'] = convert_word_to_char_indices(
            context, entry['speaker_position']
        )

    # 保存处理后的数据集
    output_path = file_path.replace("_raw", "")
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f"处理后的数据集已保存到: {output_path}")

# 使用示例
file_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\processed_val.json"
process_dataset(file_path)
```

```python
import json
import uuid  # For generating unique IDs

def add_ids_to_dataset(input_path, output_path):
    """Convert string IDs to integers using hash and save the updated dataset."""
    with open(input_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    for entry in data:
        # Convert the string ID to a numerical value (e.g., by hashing)
        if 'id' in entry:
            entry['id'] = int(uuid.UUID(entry['id']).int % (2**32))  # Ensure it's an int32-compatible value

    # Save the updated dataset with numeric IDs
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f"Updated dataset with numeric IDs saved to: {output_path}")

if __name__ == "__main__":
    input_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\test.json"
    output_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\test.json"
    
    add_ids_to_dataset(input_path, output_path)
```

```python
import torch

def process_and_save_embeddings(pt_file_path, output_file_path):
    """
    Reads data from a .pt file with multiple entries, removes entries with empty speaker positions,
    finds the closest speaker position to the target position, applies masking, and saves the updated entries.

    Args:
        pt_file_path (str): Path to the input .pt file containing individual entries.
        output_file_path (str): Path to save the updated data with new embeddings.
    """
    # Load the data from the .pt file (expected to be a list of entries)
    data_list = torch.load(pt_file_path)

    # Filter out entries with empty speaker positions
    filtered_data_list = []
    for entry_idx, entry in enumerate(data_list):
        if not entry['speaker_position']:
            print(f"Removed entry {entry_idx} due to empty speaker positions.")
            continue  # Skip this entry

        # Extract relevant fields
        token_embeddings = entry['token_embeddings']  # Shape: (seq_len, embedding_dim)
        target_position = entry['target_position']    # A single (start, end) tuple
        speaker_position = entry['speaker_position']  # List of (start, end) tuples

        seq_len, embedding_dim = token_embeddings.size()

        # Initialize speaker and target label masks
        speaker_labels = torch.zeros(seq_len)
        target_labels = torch.zeros(seq_len)

        # Calculate the center of the target position
        target_center = (target_position[0] + target_position[1]) / 2

        # Find the closest speaker position to the target position
        closest_speaker = min(
            speaker_position, 
            key=lambda pos: abs((pos[0] + pos[1]) / 2 - target_center)
        )

        # Use only the closest speaker position to generate speaker labels
        start, end = closest_speaker
        if 0 <= start < seq_len and 0 <= end < seq_len and end - start + 1 <= 500:
            speaker_labels[start:end + 1] = 1  # Mark the closest speaker positions

        # Generate target labels for the given (start, end) range
        t_start, t_end = target_position
        if 0 <= t_start < seq_len and 0 <= t_end < seq_len:
            target_labels[t_start:t_end + 1] = 1  # Mark the target range

        # Apply mask to extract target embeddings
        target_embeddings = token_embeddings * target_labels.unsqueeze(-1)

        # Add the new labels and embeddings to the entry
        entry['speaker_labels'] = speaker_labels
        entry['speaker_position'] = closest_speaker
        entry['target_embeddings'] = target_embeddings

        # Add the processed entry to the filtered list
        filtered_data_list.append(entry)

    # Save the updated data back to a new .pt file
    torch.save(filtered_data_list, output_file_path)

    print(f"Updated data with embeddings saved to: {output_file_path}")

# Example usage
pt_file_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\val_embeddings_multi.pt"
output_file_path = r"C:\Users\Lenovo\OneDrive\NUS\CS-24fall\project\AudiobookGeneration_cs5647\LiteraryTextsDataset\dataset\val_embeddings.pt"

process_and_save_embeddings(pt_file_path, output_file_path)
```
